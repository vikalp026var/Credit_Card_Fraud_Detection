```python
X=df.drop(labels=["default payment next month"],axis=1)
```

```python
y=df['default payment next month']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=50)
```

```python
X_train
```

```python
columns_to_drop=['Sex','Education','Marriage','Age']
```

```python
# if all(col in df.columns for col in columns_to_drop):
X_train_new= X_train.drop(columns=columns_to_drop, axis=1)
```

```python
##Preprocessing steps 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
train_scale=StandardScaler()
```

```python
test_scale=StandardScaler()
```

```python
X_train_scale=train_scale.fit_transform(X_train)
```

```python
X_test_scale=train_scale.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf.fit(X_train_scale,y_train)
```

```python
y_pred=clf.predict(X_test_scale)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={"var_smoothing":[1e-9,0.1,0.001,0.5,0.05,0.01,1e-8,1e-7,1e-6,1e-10,1e-11]}
```

```python
cl=GridSearchCV(clf,param_grid=param_grid,verbose=3,cv=10)
```

```python
cl.fit(X_train_scale,y_train)
```

```python
cl.best_params_
```

```python
y_test_pred=cl.predict(X_test_scale)
```

```python
print(accuracy_score(y_test,y_test_pred))
```
